```python
### Undrsampling

# build a sample dataset containing similar distribution of normal transaction and fraudlent transaction
```

```python
# no of fraudlent transaction --> 492
```

```python
legit_sample = legit.sample(n = 492)
```

```python
### concate 2 DataFrame
```

```python
new_df = pd.concat([legit_sample,fraud], axis = 0)
```

```python
new_df.head()
```

```python
new_df.Class.value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
#### Splitting the data into input & output variable
```

```python
x = new_df.iloc[:,:-1]
```

```python
y = new_df.iloc[:,-1]
```

```python
## splitng the data into train and test
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2,stratify = y, random_state = 2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
## Model training
```

```python
## Logistic Regression
```

```python
logistic_model = LogisticRegression()
```

```python
logistic_model.fit(x_train, y_train)
```

```python
### Model Evaluation 
```

```python
## accuracy_score
```

```python
x_train_prediction = logistic_model.predict(x_train)
train_accuracy = accuracy_score(x_train_prediction, y_train)*100
```

```python
train_accuracy
```

```python
## accuracy on test data
```

```python
x_test_prediction = logistic_model.predict(x_test)
test_accuracy = accuracy_score(x_test_prediction, y_test)*100
```
